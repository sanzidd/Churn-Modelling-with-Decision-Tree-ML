```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
data=pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data=data.drop(columns=['RowNumber','CustomerId','Surname'],axis=1)
```

```python
data.head()
```

```python
data.shape
```

ENCODING

```python
le_gender= LabelEncoder()
```

```python
data['Gender'] = le_gender.fit_transform(data['Gender'])
```

```python
data.head()
```

```python
ohe_geography = OneHotEncoder(drop='first')
geo_encoded =ohe_geography.fit_transform(data[['Geography']]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded,columns=ohe_geography.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
data=pd.concat([data.drop(columns=['Geography']),geo_encoded_df],axis=1)
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
exited_counts= data['Exited'].value_counts()
```

```python
plt.figure(figsize=(6,4))
plt.pie(exited_counts,
        labels=exited_counts.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=['blue','red'],
        explode=(0,0.1),
        shadow= True)
plt.title('Distribution of Exited Customers')
plt.axis('equal')
plt.show()
```

```python
data.corr()
```

```python
data.describe()
```

```python
x=data.drop(columns=['Exited'],axis=1)
```

```python
y=data[['Exited']]
```

```python
x.head()
y.head()
```

```python
x.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
cm_df = pd.DataFrame(cm, index=['Actual Negative', 'Actual Positive'], 
                     columns=['Predicted Negative', 'Predicted Positive'])

fig = px.imshow(cm_df, 
                text_auto=True, 
                color_continuous_scale="Viridis", 
                title="Confusion Matrix")

fig.update_layout(
    title={
        'text': "Confusion Matrix",
        'x': 0.5, 
        'xanchor': 'center' 
    },
    
    xaxis_title="Predicted Label",
    yaxis_title="True Label",
    coloraxis_showscale=True
)
fig.show()
```
